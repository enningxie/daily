## matplotlib.pyplot快速绘图

pyplot模块虽然用法简单，但不适合在较大的应用程序中使用。
可以在Ipython中输入类似`plt.plot??`的命令查看pyplot模块的函数是如何对各种绘图对象进行包装的。

matplotlib实际上是一套面向对象的绘图库，可以直接获取对象的属性。

matplotlib里的常用类的包含关系为Figure -> Axes -> (Line2D, Text, etc.) 一个Figure对象可以包含多个子图(Axes 该对象表示一个绘图区域，可以理解为子图)。

```python
import numpy as np
import matplotlib.pyplot as plt
 
plt.figure(1) # 创建图表1
plt.figure(2) # 创建图表2
ax1 = plt.subplot(211) # 在图表2中创建子图1
ax2 = plt.subplot(212) # 在图表2中创建子图2
 
x = np.linspace(0, 3, 100)
for i in xrange(5):
    plt.figure(1)  #❶ # 选择图表1
    plt.plot(x, np.exp(i*x/3))
    plt.sca(ax1)   #❷ # 选择图表2的子图1
    plt.plot(x, np.sin(i*x))
    plt.sca(ax2)  # 选择图表2的子图2
    plt.plot(x, np.cos(i*x))
 
plt.show()
```
